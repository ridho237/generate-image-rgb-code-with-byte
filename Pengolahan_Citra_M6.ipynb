#1. Importing Python Packages

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

# 2. Membuka gambar

```python
img1 = Image.open('/content/Naruto.jpg').convert('L')
img2 = Image.open('/content/Naruto.jpg').convert('L')

# Ubah gambar ke array NumPy untuk operasi bitwise dan aritmatika
arr1 = np.array(img1)
arr2 = np.array(img2)
```

#3. Operasi Bitwise

```python
# Bitwise AND
bitwise_and = np.bitwise_and(arr1, arr2)

# Bitwise OR
bitwise_or = np.bitwise_or(arr1, arr2)

# Bitwise XOR
bitwise_xor = np.bitwise_xor(arr1, arr2)

# Bitwise NAND
bitwise_nand = np.bitwise_not(np.bitwise_and(arr1, arr2))

# Bitwise NOR
bitwise_nor = np.bitwise_not(np.bitwise_or(arr1, arr2))

# Bitwise XNOR
bitwise_xnor = np.bitwise_not(np.bitwise_xor(arr1, arr2))
```

#4. Operasi Aritmatika

```python
# Penjumlahan (+)
add = np.clip(arr1 + arr2, 0, 255)

# Pengurangan (-)
subtract = np.clip(arr1 - arr2, 0, 255)

# Perkalian (*)
multiply = np.clip(arr1 * arr2, 0, 255)

# Pembagian (/)
# Menangani pembagian oleh nol dengan mengganti hasilnya menjadi nol
divide = np.divide(arr1, arr2, out=np.zeros_like(arr1, dtype=float), where=(arr2!=0))
divide = np.clip(divide * 255, 0, 255).astype(np.uint8)  # Normalisasi hasil ke 0-255
```

#5. Menampilkan Hasil Operasi

```python
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

# Tampilkan hasil bitwise
axs[0, 0].imshow(bitwise_and, cmap='gray')
axs[0, 0].set_title('AND')
axs[0, 1].imshow(bitwise_or, cmap='gray')
axs[0, 1].set_title('OR')
axs[0, 2].imshow(bitwise_xor, cmap='gray')
axs[0, 2].set_title('XOR')

# Tampilkan hasil bitwise NOT-based
axs[1, 0].imshow(bitwise_nand, cmap='gray')
axs[1, 0].set_title('NAND')
axs[1, 1].imshow(bitwise_nor, cmap='gray')
axs[1, 1].set_title('NOR')
axs[1, 2].imshow(bitwise_xnor, cmap='gray')
axs[1, 2].set_title('XNOR')

# Tampilkan hasil operasi aritmatika
axs[2, 0].imshow(add, cmap='gray')
axs[2, 0].set_title('Addition')
axs[2, 1].imshow(subtract, cmap='gray')
axs[2, 1].set_title('Subtraction')
axs[2, 2].imshow(multiply, cmap='gray')
axs[2, 2].set_title('Multiplication')

plt.tight_layout()
plt.show()
```
